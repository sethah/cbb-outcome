```python
import pandas as pd
from cbb_analysis import Analysis
from queries import Query
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.grid_search import GridSearchCV
%matplotlib inline
```

```python
a = Analysis()
a.convert_dates()
a.df['season'] = a.df['dt'].map(lambda x: x.year if (x.month < 6) else (x.year+1))
# drop columns we don't need
a.df = a.df.drop(['home_first', 'home_second', 'away_first', 'away_second', 
                  'box_link', 'pbp_link', 'neutral_site', 
                  'venue', 'city', 'attendance'], 1)
q = Query()
q.query = '''SELECT * FROM raw_teams'''
teams = pd.read_sql(q.query, q.conn)
box_stats = pd.read_sql("""SELECT * FROM box_stats_ss""", q.conn)
q.conn.close()

a.df = pd.merge(a.df, teams[['statsheet', 'ncaaid']], left_on='home_team', right_on='ncaaid')
a.df = pd.merge(a.df, teams[['statsheet', 'ncaaid']], left_on='away_team', right_on='ncaaid', suffixes=('_home', '_away'))
# a.df = a.df.drop(['ncaaid_home', 'ncaaid_away', 'home_team', 'away_team'], 1)
cols = list(a.df)
cols.insert(1, cols.pop(cols.index('statsheet_home')))
cols.insert(1, cols.pop(cols.index('statsheet_away')))
a.df = a.df.ix[:, cols]
```

```python
# not sure why some of the ftr are null, but drop them for now
a.df = a.df[~a.df['home_adj_ftr'].isnull()]
```

```python

# get all the rows that have a spread
a.df = a.df[~a.df['home_spread'].isnull()]
a.df['covered'] = (a.df['margin'] > -1*a.df['home_spread'])
a.df.head()
```

```python
print 'Home teams covered the spread %0.2f of the time' % (a.df.covered.sum() / float(a.df.covered.count()))
```

```python
xcols = ['home_adj_ppp', 'home_adj_dppp','away_adj_ppp', 'away_adj_dppp',
         'home_adj_efg', 'home_adj_defg','away_adj_efg', 'away_adj_defg',
         'home_adj_trt', 'home_adj_dtrt','away_adj_trt', 'away_adj_dtrt',
         'home_adj_ftr', 'home_adj_dftr','away_adj_ftr', 'away_adj_dftr', 'home_spread']
X = a.df[xcols].values
y = a.df['margin'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
spreads_train = X_train[:,-1]
spreads_test = X_test[:,-1]
X_train = np.delete(X_train, -1, 1)
X_test = np.delete(X_test, -1, 1)
```

```python
lin = LinearRegression()
lin.fit(X_train, y_train)
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
lin_pred = lin.predict(X_test)
sns.distplot(rf_pred - y_test, label='RF')
sns.distplot(lin_pred - y_test, label='Linear')
plt.legend()
```

```python
np.sum((spreads_test*(-1) < rf_pred))/float(len((spreads_test*(-1) < rf_pred)))
```

```python
xcols = ['home_adj_ppp', 'home_adj_dppp','away_adj_ppp', 'away_adj_dppp',
         'home_adj_efg', 'home_adj_defg','away_adj_efg', 'away_adj_defg',
         'home_adj_trt', 'home_adj_dtrt','away_adj_trt', 'away_adj_dtrt',
         'home_adj_ftr', 'home_adj_dftr','away_adj_ftr', 'away_adj_dftr',
         'home_spread', 'gameid']
# xcols = ['home_adj_ppp', 'home_adj_dppp','away_adj_ppp', 'away_adj_dppp', 'home_spread','season']
X = a.df[xcols].values
y = a.df['covered'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
gameid_train = X_train[:,-1]
gameid_test = X_test[:,-1]
X_train = np.delete(X_train, -1, 1)
X_test = np.delete(X_test, -1, 1)
X_train = scale(X_train)
X_test = scale(X_test)
```

```python
models = [LogisticRegression()]#, RandomForestClassifier(), GradientBoostingClassifier(), AdaBoostClassifier()]
d = {}
for model in models:
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    predicted = model.predict_proba(X_test)
    d[model.__class__.__name__] = {'score': score, 'predict': predicted[:,1]}
print d
    
```

```python
gid = pd.DataFrame({'gameid': gameid_test})
```

```python
# best parameters: {'bootstrap': True,
# 'min_samples_leaf': 4, 
# 'n_estimators': 40, 
# 'min_samples_split': 1,
# 'random_state': 1,
# 'max_features': None, 
# 'max_depth': None}
```

```python
# gid['Logistic'] = pd.Series(d['LogisticRegression']['predict'])
for model in d:
    gid[model] = pd.Series(d[model]['predict'])
```

```python
results = pd.merge(gid, a.df, on='gameid')
```

```python
mcols = [model for model in d]
other_cols = ['covered','dt','statsheet_home','statsheet_away','home_score','away_score','home_spread']
# gid = gid.drop(['home_team', 'away_team', 'home_outcome'])
results = results[mcols+other_cols]
```

```python
# mcols.pop()
print mcols
# results['agg'] = results[mcols].sum(axis=1) / len(mcols)
results['agg'] = np.where(results['LogisticRegression'] > 0.5, True, False )
results['count'] = np.where(True, True, False)

# results = results[(results['statsheet_home'] == 'indiana') | (results['statsheet_away'] == 'indiana')]
results = results[results['dt'] > pd.to_datetime('12-01-2013')]
results.sort('LogisticRegression',ascending=False).head(100)
```
